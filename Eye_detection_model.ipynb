```python
import shutil
import os
 
src = '/content/dataset/labels/'
dest = '/content/dataset/images/'

files = os.listdir(src)
 
for f in files:
    shutil.move(src + f, dest)
```

```python
%cd /
from glob import glob
from sklearn.model_selection import train_test_split

img_list = glob('content/dataset/images/*.jpg')
train_img_list, val_img_list = train_test_split(img_list, random_state=2000, test_size=0.2)
print(len(train_img_list), len(val_img_list))
```

```python
with open('/content/dataset/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list)+'\n')

with open('/content/dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list)+'\n')
```

```python
import yaml

with open('/content/dataset/data.yaml', 'r') as f:
  data = yaml.full_load(f)

print(data)

data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'

with open('/content/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
